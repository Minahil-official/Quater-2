 # **Video Generation**


## **Prompt**:
### A Fast-Paced Journey Through Human Resilience and Progress

 The video opens with a breathtaking sunrise over a bustling city, symbolizing the dawn of a new era. The camera pans to a group of children in a rural classroom, captivated by a teacher demonstrating a solar-powered device. As the camera zooms into the device, it transitions into an advanced satellite orbiting Earth, highlighting human ingenuity.

The scene shifts to a montage of people innovating in everyday life: a farmer using drones to monitor crops, a young coder designing an app in a vibrant co-working space, and an artisan creating 3D-printed prosthetics for children. Each clip bursts with energy, showcasing the creativity and determination that fuels progress.

Midway, the focus pivots to overcoming challenges: engineers reinforcing flood defenses, scientists unveiling a vaccine in a dramatic reveal, and communities rebuilding homes after a disaster. The transitions become faster, symbolizing the rapid pace of innovation driven by necessity.

In the final act, the video celebrates human connections: a family enjoying clean water thanks to innovative purification systems, diverse groups engaging in virtual reality to collaborate globally, and artists using AI to bridge cultural divides. The culmination is a grand aerial view of a city at night, lights forming a glowing network that represents unity.

The text overlay reads: "Progress thrives on resilience. The future is ours to build."



```python
Generated_video = "/content/Unstoppable Human Spirit_ Innovations Unleashed!-VEED.mp4"
```

```python
import time

def upload_video(Generated_video):
  video_file = client.files.upload(path=Generated_video)
  while video_file.state == "PROCESSING":
      print('Waiting for video to be processed.')
      time.sleep(10)
      video_file = client.files.get(name=video_file.name or "")

  if video_file.state == "FAILED":
    raise ValueError(video_file.state)
  print(f'Video processing complete: ' + (video_file.uri or ""))

  return video_file

video_script = upload_video(Generated_video)
```

 # **Video Script Generation**

```python
from google.genai.types import Content, Part
prompt = """For each scene in this video,
            generate captions that describe the scene along with any spoken text placed in quotation marks.
            Place each caption into an object with the timecode of the caption in the video.
         """

video = video_script

response = client.models.generate_content(
    model=model,
    contents=[
        Content(
            role="user",
            parts=[
                Part.from_uri(
                    file_uri=video.uri or "",
                    mime_type=video.mime_type or ""),
                ]),
        prompt,
    ]
)
print(response.text)
```

# **Extracting Audio**

```python
import json
from gtts import gTTS
from IPython.display import Audio, display
from pydub import AudioSegment

# JSON text
response_text = '''
[
    {
        "timecode": "00:00",
        "caption": "A wide shot of a calm body of water with trees and vegetation on the far bank. The sun shines brightly in the sky, and text appears over an image of a sun saying, 'EVERY SUNRISE'."
    },
    {
        "timecode": "00:01",
        "caption": "The text changes to read, 'EVERY SUNRISE MARKS A NEW BEGINNING,' over the water."
    },
    {
        "timecode": "00:02",
        "caption": "A floating die appears with the text reading, 'A CHANCE TO INNOVATE AND INSPIRE'."
    },
    {
        "timecode": "00:04",
        "caption": "A group of young people walk along a paved road in a rural area wearing school uniforms with backpacks. Text that says, 'IN RURAL,' appears over a sign with a stick figure of a child."
    },
    {
        "timecode": "00:06",
        "caption": "Text that reads, 'IN RURAL CLASSROOMS, CHILDREN ARE LEARNING' appears over the students walking to school."
    },
    {
        "timecode": "00:07",
        "caption": "Text appears reading, 'ABOUT SOLAR POWER, IGNITING THEIR IMAGINATIONS.'"
    },
    {
        "timecode": "00:10",
        "caption": "Inside a spacecraft, an astronaut wears white gloves and operates a control panel with screens. Text appears reading, 'MEANWHILE,'"
    },
    {
        "timecode": "00:11",
        "caption": "Text appears reading, 'FARMERS ARE HARNESSING DRONES.'"
    },
    {
        "timecode": "00:13",
        "caption": "The text on the screen reads, 'REVOLUTIONIZE AGRICULTURE, AND YOUNG'."
    },
    {
        "timecode": "00:16",
        "caption": "The text changes to read, 'CRAFTING APPS THAT CONNECT US ALL.'"
    },
    {
        "timecode": "00:18",
        "caption": "A young man sits at a desk and works on a laptop. The text reads, 'BUT PROGRESS'."
    },
    {
        "timecode": "00:20",
        "caption": "The text changes to read, 'BUT PROGRESS ISN'T WITHOUT ITS HURDLES.'"
    },
    {
        "timecode": "00:21",
        "caption": "An aerial view of construction machinery working in a muddy pit, with text reading, 'ENGINEERS ARE'."
    },
    {
        "timecode": "00:22",
        "caption": "The text changes to read, 'ENGINEERS ARE FORTIFYING FLOOD DEFENSES.'"
    },
    {
        "timecode": "00:24",
        "caption": "Two scientists wear lab coats and safety glasses and hold up a test tube with red liquid to the light with the text reading, 'SCIENTISTS'."
    },
    {
        "timecode": "00:26",
        "caption": "The text changes to read, 'SCIENTISTS UNVEIL LIFE SAVING VACCINES.'"
    }
]
'''

# Parse the JSON
data = json.loads(response_text)

# Initialize an empty audio segment
combined_audio = AudioSegment.silent(duration=0)

# Process each caption
for item in data:
    caption = item["caption"]  # Extract caption text
    print(f"Generating audio for: {caption}")

    # Convert text to speech using gTTS
    tts = gTTS(text=caption, lang='en')
    filename = f"{item['timecode'].replace(':', '_')}.mp3"  # Safe filename
    tts.save(filename)  # Save the audio

    # Load the generated audio
    audio_segment = AudioSegment.from_file(filename)
    combined_audio += audio_segment  # Add to the combined audio

# Export combined audio
output_file = "combined_output.mp3"
combined_audio.export(output_file, format="mp3")

# Play the combined audio
print("Playing combined audio with captions...")
display(Audio(output_file, autoplay=True))
```
